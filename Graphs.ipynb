```python
import networkx as nx
import matplotlib.pyplot as plt



G = nx.Graph()
edges = nx.read_edgelist("C:\\Users\\Abdul\\Desktop\\influenced by1.txt")

options = {
    'node_color': 'blue',
    'node_size': 50,
    'line_color': 'black',
    'linewidths': 0,
    'width': 0.1,
}

print(nx.info(edges))

print("Clustering",nx.average_clustering(edges))
pos = nx.spring_layout(edges,k=0.7,iterations=200)

nx.draw(edges,pos=pos,**options,with_labels=True,font_size=5)


fig = plt.gcf()
fig.set_size_inches((10,10))

fig.suptitle('Programming Languages Influenced by Other Languages',size = 10,color='black')
fig = plt.gcf()


#fig.patch.set_facecolor('black')
plt.show()
```

```python

def plot_degree_distribution (wiki) :
    degs = {}
    for n in wiki.nodes () :
        deg = wiki.degree ( n )
        
        
        if deg not in degs :
            degs [ deg ] = 0
        
        degs [ deg ] += 1
    items = sorted ( degs.items () )
            
    fig = plt.figure ()
    ax = fig.add_subplot (111)
    ax.plot ([ k for (k , v ) in items ] , [ v for (k ,v ) in items ])
    ax.set_xscale('log')
    ax.set_yscale('log')
    plt.title("Degree Distribution Of Languages Influenced by")
    
    
            
plot_degree_distribution(edges)
```
